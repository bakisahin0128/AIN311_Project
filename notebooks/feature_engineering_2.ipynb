```python
import pandas as pd
import numpy as np

# Veri setini yükleyin
input_file_path = r"C:\Users\mbaki\Desktop\Proje\data\processed\final_prepared_data.csv"
df = pd.read_csv(input_file_path)

# Maçları kronolojik sıraya göre sıralayın
df = df.sort_values('Match_Number').reset_index(drop=True)

# Takım geçmişini saklamak için sözlük oluşturun
team_history = {}

# Özellik listeleri (son 5 maç)
home_avg_goals_last5 = []
away_avg_goals_last5 = []
home_wins_last5 = []
away_wins_last5 = []
home_avg_performance_last5 = []
away_avg_performance_last5 = []

# Özellik listeleri (son 10 maç)
home_avg_goals_last10 = []
away_avg_goals_last10 = []
home_wins_last10 = []
away_wins_last10 = []
home_avg_performance_last10 = []
away_avg_performance_last10 = []

# Ev sahibi avantajını ekleyin
df['Home_Advantage'] = 1  # Her maçta ev sahibi için 1

# Oyuncu skorlarının ortalamasını ekleyin
home_player_cols = [f'Home Player {i}' for i in range(1, 12)]
away_player_cols = [f'Away Player {i}' for i in range(1, 12)]
df['Home_Avg_Player_Score'] = df[home_player_cols].mean(axis=1)
df['Away_Avg_Player_Score'] = df[away_player_cols].mean(axis=1)

# Takım geçmişini güncelleyen yardımcı fonksiyon
def update_team_history(history, match_result, goals_scored, performance):
    """
    Takım geçmişini günceller.
    """
    history.append({
        'Match_Result': match_result,
        'Goals_Scored': goals_scored,
        'Performance': performance
    })
    # Sadece son 10 maçı saklamak için
    if len(history) > 10:
        history.pop(0)

# Maçları işleyerek özellikleri hesaplayın
for idx, row in df.iterrows():
    home_team = row['Home Team']
    away_team = row['Away Team']
    
    # Ev sahibi takımın geçmişini al veya oluştur
    if home_team not in team_history:
        team_history[home_team] = []
    home_history = team_history[home_team]
    
    # Deplasman takımının geçmişini al veya oluştur
    if away_team not in team_history:
        team_history[away_team] = []
    away_history = team_history[away_team]
    
    # Ev sahibi takım için son 5 maçın ortalama gol sayısı
    if len(home_history) >= 1:
        avg_goals_5 = np.mean([match['Goals_Scored'] for match in home_history[-5:]])
    else:
        avg_goals_5 = 0
    home_avg_goals_last5.append(avg_goals_5)
    
    # Deplasman takım için son 5 maçın ortalama gol sayısı
    if len(away_history) >= 1:
        avg_goals_5 = np.mean([match['Goals_Scored'] for match in away_history[-5:]])
    else:
        avg_goals_5 = 0
    away_avg_goals_last5.append(avg_goals_5)
    
    # Ev sahibi takım için son 5 maçta galibiyet sayısı
    if len(home_history) >= 1:
        wins_5 = sum([1 for match in home_history[-5:] if match['Match_Result'] == 1])
    else:
        wins_5 = 0
    home_wins_last5.append(wins_5)
    
    # Deplasman takım için son 5 maçta galibiyet sayısı
    if len(away_history) >= 1:
        wins_5 = sum([1 for match in away_history[-5:] if match['Match_Result'] == 1])
    else:
        wins_5 = 0
    away_wins_last5.append(wins_5)
    
    # Ev sahibi takım için son 5 maçın ortalama performansı
    if len(home_history) >= 1:
        avg_perf_5 = np.mean([match['Performance'] for match in home_history[-5:]])
    else:
        avg_perf_5 = 0
    home_avg_performance_last5.append(avg_perf_5)
    
    # Deplasman takım için son 5 maçın ortalama performansı
    if len(away_history) >= 1:
        avg_perf_5 = np.mean([match['Performance'] for match in away_history[-5:]])
    else:
        avg_perf_5 = 0
    away_avg_performance_last5.append(avg_perf_5)
    
    # Ev sahibi takım için son 10 maçın ortalama gol sayısı
    if len(home_history) >= 1:
        avg_goals_10 = np.mean([match['Goals_Scored'] for match in home_history[-10:]])
    else:
        avg_goals_10 = 0
    home_avg_goals_last10.append(avg_goals_10)
    
    # Deplasman takım için son 10 maçın ortalama gol sayısı
    if len(away_history) >= 1:
        avg_goals_10 = np.mean([match['Goals_Scored'] for match in away_history[-10:]])
    else:
        avg_goals_10 = 0
    away_avg_goals_last10.append(avg_goals_10)
    
    # Ev sahibi takım için son 10 maçta galibiyet sayısı
    if len(home_history) >= 1:
        wins_10 = sum([1 for match in home_history[-10:] if match['Match_Result'] == 1])
    else:
        wins_10 = 0
    home_wins_last10.append(wins_10)
    
    # Deplasman takım için son 10 maçta galibiyet sayısı
    if len(away_history) >= 1:
        wins_10 = sum([1 for match in away_history[-10:] if match['Match_Result'] == 1])
    else:
        wins_10 = 0
    away_wins_last10.append(wins_10)
    
    # Ev sahibi takım için son 10 maçın ortalama performansı
    if len(home_history) >= 1:
        avg_perf_10 = np.mean([match['Performance'] for match in home_history[-10:]])
    else:
        avg_perf_10 = 0
    home_avg_performance_last10.append(avg_perf_10)
    
    # Deplasman takım için son 10 maçın ortalama performansı
    if len(away_history) >= 1:
        avg_perf_10 = np.mean([match['Performance'] for match in away_history[-10:]])
    else:
        avg_perf_10 = 0
    away_avg_performance_last10.append(avg_perf_10)
    
    # Takım geçmişini güncelleyin
    # Maç sonucunu sayısal olarak varsayalım: 1 = Galibiyet, 0 = Beraberlik, -1 = Mağlubiyet
    update_team_history(
        home_history,
        row['Match_Result'],  # Ev sahibi takımın sonucu
        row['Home Goals'],
        row['Home Performance']
    )
    
    update_team_history(
        away_history,
        row['Match_Result'],  # Deplasman takımının sonucu
        row['Away Goals'],
        row['Away Performance']
    )

# Yeni özellikleri DataFrame'e ekleyin
df['Home_Avg_Goals_Last5'] = home_avg_goals_last5
df['Away_Avg_Goals_Last5'] = away_avg_goals_last5
df['Home_Wins_Last5'] = home_wins_last5
df['Away_Wins_Last5'] = away_wins_last5
df['Home_Avg_Performance_Last5'] = home_avg_performance_last5
df['Away_Avg_Performance_Last5'] = away_avg_performance_last5

df['Home_Avg_Goals_Last10'] = home_avg_goals_last10
df['Away_Avg_Goals_Last10'] = away_avg_goals_last10
df['Home_Wins_Last10'] = home_wins_last10
df['Away_Wins_Last10'] = away_wins_last10
df['Home_Avg_Performance_Last10'] = home_avg_performance_last10
df['Away_Avg_Performance_Last10'] = away_avg_performance_last10

# Performans farklarını hesaplayın
df['Performance_Diff_Last5'] = df['Home_Avg_Performance_Last5'] - df['Away_Avg_Performance_Last5']
df['Performance_Diff_Last10'] = df['Home_Avg_Performance_Last10'] - df['Away_Avg_Performance_Last10']

# Dummy değişkenler oluşturmak yerine formasyon sütunlarını kaldırın
df = df.drop(['Home Formation', 'Away Formation'], axis=1)

# Formasyon çeşitliliği özelliklerini kaldırın (isteğe bağlı)
# Eğer formasyon çeşitliliği özelliklerini eklememişseniz bu adımı atlayabilirsiniz.

# Eksik değerleri doldurun
df.fillna(0, inplace=True)

# Diğer kategorik değişkenler için dummy değişkenler oluşturun (varsa)
# Örneğin, takım isimleri veya diğer kategorik değişkenler
# Ancak, takım isimleri genellikle modellemeye dahil edilmez. Eğer dahil etmek isterseniz, uygun şekilde encode etmelisiniz.

# Hedef değişkenleri ayırın
X = df.drop(['Match_Result', 'Goal_Difference'], axis=1)
y_result = df['Match_Result']
y_goals = df['Goal_Difference']

# Hazırlanan DataFrame'i yeni bir CSV dosyasına kaydedin
output_file_path = r"C:\Users\mbaki\Desktop\Proje\data\processed\final_prepared_data_with_features.csv"
df.to_csv(output_file_path, index=False)

print(f"Veri seti başarıyla '{output_file_path}' olarak kaydedildi.")
```
