```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Load CSV Files
csv1 = pd.read_csv(r"C:\Users\mbaki\Desktop\Proje\data\raw\23_24\23_24.csv")
csv2 = pd.read_csv(r"C:\Users\mbaki\Desktop\Proje\data\raw\23_24\23_24_teams_and_players.csv")
csv3 = pd.read_csv(r"C:\Users\mbaki\Desktop\Proje\data\raw\23_24\23_24_marketvalue_and_age.csv")
```

```python
# Detect rows containing '-'
mask = (csv3['Age'] == '-') | (csv3['Market Value'] == '-')

# Set 'Age' and 'Market Value' to NaN in these rows
csv3.loc[mask, ['Age', 'Market Value']] = np.nan

# Convert 'Age' column to numeric (NaN values are automatically preserved)
csv3['Age'] = pd.to_numeric(csv3['Age'], errors='coerce')

# Clean and convert 'Market Value' column to numeric
# First, convert expressions like 'mil. €' and 'bin €' to numerical values
def convert_market_value(value):
    if pd.isna(value):
        return np.nan
    value = value.replace(' mil. €', 'e6').replace(' bin €', 'e3').replace('€', '').replace('.', '').replace(',', '.')
    try:
        return float(value)
    except:
        return np.nan

csv3['Market Value'] = csv3['Market Value'].apply(convert_market_value)

# Check data types after cleaning
print("\nAfter Preprocessing:")
print(csv3.dtypes)

# Check the number of NaN values
print("\nNaN Values:")
print(csv3.isna().sum())
```

```python
# Merge two CSV files
merged_csv2_csv3 = pd.merge(csv2, csv3, on=['Team Name', 'Player Name'], how='left')

# Check the merged data
print(merged_csv2_csv3.head())
```

```python
# Fill NaN values with team-wise averages
merged_csv2_csv3['Age'] = merged_csv2_csv3.groupby('Team Name')['Age'].transform(lambda x: x.fillna(x.mean()))
merged_csv2_csv3['Market Value'] = merged_csv2_csv3.groupby('Team Name')['Market Value'].transform(lambda x: x.fillna(x.mean()))

# Verify the results
print(merged_csv2_csv3.head())
```

```python
def extract_surname(full_name):
    name = full_name.replace('(c)', '').strip()
    parts = name.split(' ')
    return parts[-1] if parts else name

# Function to modify team names
def modify_team_name(team_name):
    team_name_lower = team_name.lower()
    if team_name_lower == 'karagümrük':
        return 'fati'
    elif team_name_lower == 'mke ankaragücü':
        return 'anka'
    elif team_name_lower == 'çaykur rizespor':
        return 'rize'
    else:
        return team_name_lower[:4]
```

```python
# Split player names into lists
csv1['Home Players'] = csv1['Home Players'].str.split('; ')
csv1['Away Players'] = csv1['Away Players'].str.split('; ')

# Extract surnames for Home and Away players
csv1['Home Players'] = csv1['Home Players'].apply(lambda players: [extract_surname(player) for player in players])
csv1['Away Players'] = csv1['Away Players'].apply(lambda players: [extract_surname(player) for player in players])

MAX_PLAYER = 11

# Split player lists into separate columns
def split_players(player_list, prefix):
    df = player_list.apply(lambda players: pd.Series(players[:MAX_PLAYER]))
    df = df.rename(columns=lambda x: f'{prefix} Player {x + 1}')
    return df

home_players_df = split_players(csv1['Home Players'], 'Home')
away_players_df = split_players(csv1['Away Players'], 'Away')

# Remove original player columns and add new separated columns
csv1 = csv1.drop(['Home Players', 'Away Players'], axis=1)
csv1 = pd.concat([csv1, home_players_df, away_players_df], axis=1)

# Modify Home and Away Team names
csv1['Home Team'] = csv1['Home Team'].apply(modify_team_name).str.strip()
csv1['Away Team'] = csv1['Away Team'].apply(modify_team_name).str.strip()
```

```python
# Clean csv2 dataset
merged_csv2_csv3['Player Name'] = csv2['Player Name'].apply(extract_surname).str.lower().str.strip()
merged_csv2_csv3['Team Name'] = csv2['Team Name'].apply(modify_team_name).str.strip()
player_rating_dict = merged_csv2_csv3.set_index('Player Name')['Player Rating'].to_dict()
```

```python
# Combine Team and Player names in merged_csv2_csv3 and convert to lowercase
merged_csv2_csv3['Team Name'] = merged_csv2_csv3['Team Name'].str.lower().str.strip()
merged_csv2_csv3['Player Name'] = merged_csv2_csv3['Player Name'].str.lower().str.strip()

merged_csv2_csv3['Team_Player'] = merged_csv2_csv3['Team Name'] + '_' + merged_csv2_csv3['Player Name']

print(merged_csv2_csv3)
```

```python
# Function to concatenate team and player names in lowercase
def concatenate_team_player_lower(team_column, player_column):
    return team_column.str.lower().str.strip() + '_' + player_column.str.lower().str.strip()

# Concatenate Home Players
for i in range(1, MAX_PLAYER + 1):
    home_player_col = f'Home Player {i}'
    new_col = f'Home_Player_{i}_TeamPlayer'
    csv1[new_col] = concatenate_team_player_lower(csv1['Home Team'], csv1[home_player_col])

# Concatenate Away Players
for i in range(1, MAX_PLAYER + 1):
    away_player_col = f'Away Player {i}'
    new_col = f'Away_Player_{i}_TeamPlayer'
    csv1[new_col] = concatenate_team_player_lower(csv1['Away Team'], csv1[away_player_col])

# Remove original player columns after concatenation
home_player_team_cols = [f'Home Player {i}' for i in range(1, MAX_PLAYER + 1)]
away_player_team_cols = [f'Away Player {i}' for i in range(1, MAX_PLAYER + 1)]

# Collect all Team_Player columns
team_player_cols = home_player_team_cols + away_player_team_cols

# Drop existing Team_Player columns if they exist
existing_team_player_cols = [col for col in team_player_cols if col in csv1.columns]
csv1 = csv1.drop(existing_team_player_cols, axis=1)

print(csv1)
```

```python
# -------------------------------------------------------------
# 1) Select necessary columns from merged_csv2_csv3
#    based on Team_Player for Age, Market Value, Rating
# -------------------------------------------------------------
df2_subset = merged_csv2_csv3[['Team_Player', 'Age', 'Market Value', 'Player Rating']].copy()

# -------------------------------------------------------------
# 2) Get list of Home_Player_i_TeamPlayer and Away_Player_i_TeamPlayer columns
# -------------------------------------------------------------
home_teamplayer_cols = [f'Home_Player_{i}_TeamPlayer' for i in range(1, MAX_PLAYER + 1)]
away_teamplayer_cols = [f'Away_Player_{i}_TeamPlayer' for i in range(1, MAX_PLAYER + 1)]
player_teamplayer_cols = home_teamplayer_cols + away_teamplayer_cols

# -------------------------------------------------------------
# 3) Loop through each TeamPlayer column to merge Age, Market Value, Rating
#    into csv1
# -------------------------------------------------------------
for col in player_teamplayer_cols:
    # Step 3a: Merge based on Team_Player
    csv1 = pd.merge(
        csv1,
        df2_subset,
        how="left",
        left_on=col,     # csv1's Home_Player_i_TeamPlayer or Away_Player_i_TeamPlayer
        right_on="Team_Player",
        suffixes=("", f"_{col}")  # Add suffix to avoid column name conflicts
    )

    # Step 3b: Rename the merged columns appropriately
    csv1.rename(
        columns={
            'Age': f"{col}_Age",
            'Market Value': f"{col}_MarketValue",
            'Player Rating': f"{col}_Rating"
        },
        inplace=True
    )

    # Step 3c: Drop the Team_Player column after merging
    if "Team_Player" in csv1.columns:
        csv1.drop(columns="Team_Player", inplace=True)

# -------------------------------------------------------------
# 4) Verify the added columns in csv1
# -------------------------------------------------------------
print("Age, Market Value, and Player Rating columns added to csv1:\n")
print(csv1.head())
```
